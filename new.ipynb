```python
import pandas as pd
import yfinance as yf
import matplotlib as plt
import tpqoa
```

```python
start = '2019-01-01'
end = '2025-12-01'
ticker = ['BE', 'TGT', 'BA', 'EURUSD=X', 'XAUT-USD']

data = yf.download(ticker, start, end)

data.to_csv('daataset.csv')
```

```python
data.info()
```

```python
# data.head(10)
close = data['Close'].copy().dropna()
close
```

```python
normalized =close.div(close.iloc[1])

normalized.plot()
```

```python
be = normalized['BE'].dropna()
be = be.to_frame()
be['previous'] = be.shift(periods=1)

be['difference'] = be.BE - be.previous
```

```python
be['Percent_Change'] = (be.difference/be.previous) * 100
```
